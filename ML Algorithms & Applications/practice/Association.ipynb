## 1 Association Rule

```python
import pandas as pd
```

```python
df = pd.read_excel('Online_Retail.xlsx')
```

```python
df.head()
```

```python
df.shape
```

```python
basket = df[df['Country']=='France'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().fillna(0)
```

```python
basket
```

```python
df[df['Country']=='France'].groupby(['InvoiceNo','Description'])['Quantity'].sum()
```

```python
def encode(x):
    if x<=0:
        return 0
    else:
        return 1
```

```python
basket_sets = basket.applymap(encode)
```

```python
basket_sets.head()
```

```python
!pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_itemset =  apriori(basket_sets, min_support=0.04, use_colnames=True)
```

```python
rules = association_rules(frequent_itemset, metric='lift',min_threshold=1)
```

```python
rules
```

```python
rules[rules['confidence']>0.8]
```

```python
Antecedent A = item is purchased
consequent C = chance of going to be puchased

support(A) = prod A / total no. of prod
support = product a&c / total number of product = likeihood of purchasing product xy (frequently)

confidence = product A&C / support(A)

lift = support / support(A)*support(C)
```
