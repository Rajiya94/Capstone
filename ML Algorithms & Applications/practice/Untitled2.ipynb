```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Online_Retail.csv')
df.head()
```

```python
df.shape
```

```python
df[df['Country']=='France'].shape
```

```python
df[df['Country']=='France'].groupby(['InvoiceNo','Description'])['Quantity'].sum()
```

```python
basket=df[df['Country']=='France'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().fillna(0)
```

```python
basket
```

```python
def encode(x):
    if x<=0:
        return 0
    else:
        return 1
```

```python
basket_sets=basket.applymap(encode)
```

```python
basket_sets.head()
```

```python
!pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_itemset=apriori(basket_sets,min_support=0.04,use_colnames=True)
frequent_itemset.head()
```

```python
rules=association_rules(frequent_itemset,metric='lift',min_threshold=1)
rules.head()
```

```python
rules[rules['confidence']>.8]
```

1. antecedents = item is purchased
2. consequents = chance of going to purchased
3. support(A) = product A /total number of product = likelihood of purchasing product xx (frequently)
4. support(C) = product C /total number of product
5. support = product A & C/total number of product
6. confidence = product A & C /support(A)
7. lift = support / support(A)*support(C)
